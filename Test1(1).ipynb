```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys

CSV_FILE_PATH = pd.read_csv(r'/home/mrroy/python/dataset/trainLabels.csv')

IMG_FILE =r'/home/mrroy/python/dataset/eye'
CLASS = r"/home/mrroy/python/dataset/CLASS"
if not os.path.exists(CLASS):
    os.mkdir(CLASS)

for filename, class_name in CSV_FILE_PATH.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(CLASS + str(class_name)):
        os.mkdir(CLASS + str(class_name))
    src_path = IMG_FILE + '/'+ filename + '.jpeg'
    dst_path = CLASS + str(class_name) + '/' + filename + '.jpeg'
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print('Unable to copy file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import os
CLASS = r"/home/mrroy/python/dataset/CLASS"
if not os.path.exists(CLASS):
    os.mkdir(CLASS)
```

```python
batch_size = 32
img_height = 224
img_width = 224

train_data_gen = ImageDataGenerator(
    rescale=1.0/255,
    validation_split=0.2  # Split your data into training and validation sets
)

train_data = train_data_gen.flow_from_directory(
    CLASS,
    target_size=(128, 128),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'  # Use 'training' subset for training
)

validation_data = train_data_gen.flow_from_directory(
    CLASS,
    target_size=(128, 128),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'  # Use 'validation' subset for validation
)
```

```python
base_model = keras.applications.VGG16(
    weights='imagenet',
    include_top=False,  # Exclude the fully connected layers
    input_shape=(128, 128, 3)  # Adjust input shape based on your images
)
```

```python
num_classes = 5  # Replace 10 with the actual number of classes in your dataset

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),  # Add Global Average Pooling layer
    tf.keras.layers.Dense(256, activation='relu'),  # Add a dense layer
    tf.keras.layers.Dropout(0.5),  # Add dropout for regularization
    tf.keras.layers.Dense(num_classes, activation='softmax')  # Output layer for your classes
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```
