```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import random

# Define paths
CSV_FILE_PATH = pd.read_csv(r'/home/mrroy/Dataset/trainLabels.csv')
IMG_FILE = r'/home/mrroy/Dataset/eye'
CLASS = r"/home/mrroy/Dataset/CLASS"

# Define the ratios for train, test, and validation
train_ratio = 0.7
test_ratio = 0.2
validation_ratio = 0.1

# Create train, test, and validation directories in CLASS folder
for folder in ["train", "test", "validation"]:
    path = os.path.join(CLASS, folder)
    if not os.path.exists(path):
        os.makedirs(path)

# Iterate through the CSV data for each class (0, 1, 2, 3, 4)
for class_name in ["0", "1", "2", "3", "4"]:
    class_dir = os.path.join(CLASS, class_name)
    for folder in ["train", "test", "validation"]:
        path = os.path.join(CLASS, folder, class_name)
        if not os.path.exists(path):
            os.makedirs(path)

    # Iterate through the images in the current class folder
    images = os.listdir(class_dir)

    # Shuffle the list of images to randomize the order
    random.shuffle(images)

    # Determine the number of images for each split
    num_images = len(images)
    num_train = int(train_ratio * num_images)
    num_test = int(test_ratio * num_images)

    # Copy the images to their respective split folders
    for i, filename in enumerate(images):
        if i < num_train:
            folder = "train"
        elif i < num_train + num_test:
            folder = "test"
        else:
            folder = "validation"

        src_path = os.path.join(class_dir, filename)
        dst_path = os.path.join(CLASS, folder, class_name, filename)

        try:
            shutil.move(src_path, dst_path)
            print(f"Successfully copied {filename} to {folder}/{class_name}")
        except IOError as e:
            print(f'Unable to copy file {src_path} to {dst_path}')
        except Exception as e:
            print(f'When trying to copy file {src_path} to {dst_path}, unexpected error: {str(e)}')
```
